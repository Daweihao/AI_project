```python
import cv2
import os
import operator
import numpy as np
```

```python
#link: https://docs.opencv.org/3.1.0/d4/d73/tutorial_py_contours_begin.html
def getEars(img1, imgName):
     # gussian Blur
    img2 = cv2.GaussianBlur(img1,(3,3),0)
    gray1 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
   
    ret1, thresh1 = cv2.threshold(gray1, 127, 255, 0)
    contours, hierarchy = cv2.findContours(thresh1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    areas = {}
    count = 0
    for cnt in contours:
        areas[count] = cv2.contourArea(cnt)
        count+=1
        
    sort_areas = sorted(areas.items(), key=operator.itemgetter(1))
    cnts = []
    for i in sort_areas:
        if i[1] > 90000 and i[1] < 3600000:
            cnts.append(contours[i[0]])
    
    print(imgName + " " + str(len(cnts)))
    stencil = np.zeros(img1.shape).astype(img1.dtype)
    fill_color = [255, 255, 255]
    cv2.fillPoly(stencil, cnts, fill_color)
    result = cv2.bitwise_and(img1, stencil)
    result = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)
    result = cv2.equalizeHist(result)
    path = 'constrainedAfterCroped_G/'
    cv2.imwrite(os.path.join(path, imgName), result)
#     cv2.namedWindow('image', cv2.WINDOW_NORMAL)
#     gray1_rs = cv2.resize(result, (756, 1008))
#     cv2.imshow('image', gray1_rs)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
```

```python
path = 'earImageDataset/'
dirs = os.listdir(path)
for file in dirs:
    if file == '.DS_Store':
        continue
    img = cv2.imread(os.path.join(path, file))
    print(file)
    getEars(img, file)
```
